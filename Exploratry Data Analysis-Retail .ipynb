```python
# Category wise Sales
cs = df.groupby("Category").Sales.sum()

# Category wise Profit
cp = df.groupby("Category").Profit.sum()
```

```python
# Sub-Category
sub = df["Sub-Category"].value_counts()
```

```python
x = df.groupby(["Category", "Sub-Category"]).Sales.sum()
y = df.groupby(["Category", "Sub-Category"]).Profit.sum()
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(20, 20))

ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
ax4 = fig.add_subplot(2, 2, 4)

# Furniture
x["Furniture"].sort_values(ascending=True).plot(kind="barh", ax=ax1)
ax1.set_title("Furniture")
ax1.set_xlabel("Sales")
ax1.set_ylabel("Sub-Categories")

# Technology
x["Technology"].sort_values(ascending=True).plot(kind="barh", ax=ax2)
ax2.set_title("Technology")
ax2.set_xlabel("Sales")
ax2.set_ylabel("Sub-Categories")

# Office Supplies
x["Office Supplies"].sort_values(ascending=True).plot(kind="barh", ax=ax3)
ax3.set_title("Office Supplies")
ax3.set_xlabel("Sales")
ax3.set_ylabel("Sub-Categories")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(20, 20))

ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
ax0 = fig.add_subplot(2, 2, 4)

# Furniture
y["Furniture"].sort_values(ascending=True).plot(kind="barh", ax=ax1, color="red")
ax1.set_title("Furniture")
ax1.set_xlabel("Profits")
ax1.set_ylabel("Sub-Categories")

# Technology
y["Technology"].sort_values(ascending=True).plot(kind="barh", ax=ax2, color="red")
ax2.set_title("Technology")
ax2.set_xlabel("Profits")
ax2.set_ylabel("Sub-Categories")

# Office Supplies
y["Office Supplies"].sort_values(ascending=True).plot(kind="barh", ax=ax3, color="red")
ax3.set_title("Office Supplies")
ax3.set_xlabel("Profits")
ax3.set_ylabel("Sub-Categories")

# Total
y.sort_values(ascending=True).plot(kind="barh", ax=ax0, color="red")
ax0.set_title("All Sub-Categories")
ax0.set_xlabel("Profits")
ax0.set_ylabel("Sub-Categories")

plt.tight_layout()
plt.show()
```
